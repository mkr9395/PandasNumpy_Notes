### crosstab

```python
pd.crosstab(df1['Name'],df1['Type 1'])
```

```python
pd.crosstab(df1['Name'],df1['Type 1'],margins=True)
# margin gives total of the row
```

```python
pd.crosstab(df1['Name'],[df1['Type 1'],df1['Type 2']])
```

```python
pd.crosstab([df1['Type 1'],df1['Type 2']],df1['Name'])
```

```python
pd.crosstab(df1['Name'],df1['Type 1'],values=df1['Attack'],aggfunc=np.unique)
```

```python
pd.crosstab(index=[df1['Name']],columns=[df1['Type 1'],df1['Type 2']])
```
