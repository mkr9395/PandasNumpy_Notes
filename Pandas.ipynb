```python
# Solution 1

l = df1.values[df1<0]
for c in df1.columns:                 #list of all columns
    for v in df1.index:               #list of all rows
        if df1[c][v] in l:
            print(c,v,df1[c][v])
```

### MERGE OPERATION

#### Merge is done on basis of common keys.

```python
np.random.seed(23)
arr1 = np.random.randn(3,4)
df11 = pd.DataFrame(arr1,columns=['A','B','C','D'],index = [0,1,2])
df22 = pd.DataFrame(arr1,columns=['A','B','C','D'],index = [3,4,5])
```

```python
df11
```

```python
df22
```

```python
# merging both df on basis of vales of A where values are equal
pd.merge(df11,df22,on = "A")
```

```python
df1 = pd.DataFrame({'key':['K0','K8','K2','K3'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})


df2 = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on= 'key')

# inner join is happening here i.e whatever data is common between left and right on basis of common key
```

```python
# left dataframe that is df1
# consider everything from left dataframe

pd.merge(df1,df2,on= 'key',how = 'left')

# whatever is present in right but not present in left gets filled with NaN
```

```python
# right dataframe that is df2
# consider everything from right dataframe

pd.merge(df1,df2,on= 'key',how ='right')

# whatever is present in left but not present in right gets filled with NaN
```

```python
# takes all values of both df
pd.merge(df1,df2,on= 'key',how ='outer')

# it takes all the values from both the dataframe
```

### How JOIN is different from MERGE

```python
df1 = pd.DataFrame({'key1':['K0','K0','K2','K2'],
                    'key2':['K0','K1','K0','K1'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})


df2 = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                    'key2':['K0','K0','K0','K0'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
```

```python
df1
```

```python
df2
```

```python
# merge on 2 keys
pd.merge(df1,df2,on = ['key1','key2'])
```

```python
df1 = pd.DataFrame({'A':['A0','A1','A2'],
                    'B':['B0','B1','B2']},
                  index= ['K0','K1','K2'])


df2 = pd.DataFrame({'C':['C0','C2','C3'],
                    'D':['D0','D2','D3']},
                  index = ['K0','K2','K3'])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)

#left join by default
```

### NOTE:

### in Merge the dfs are merging on basis of 'columns'
### in Join the dfs are joining on basis of 'indexes'

```python
#df2 is right, df1 is left
df1.join(df2,how = 'right')
```

```python
df1.join(df2,how = 'outer')

# everything from everone, if data not present then NaN
```
